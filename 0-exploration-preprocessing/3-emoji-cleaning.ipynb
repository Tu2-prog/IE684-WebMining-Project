```python
import pandas as pd
```

```python
df = pd.read_csv("../data/cleaned_reviews_with_sentiment.csv")
df
```

```python
import emoji
def remove_emoji(text):
    return emoji.replace_emoji(text, replace='')

df["text"] = df["text"].apply(remove_emoji)
df
```

```python
df['text'] = df['text'].str.lstrip()
```

```python
df
```

```python
df.to_csv("../data/cleaned_reviews_without_emojis.csv", index=False)
```
