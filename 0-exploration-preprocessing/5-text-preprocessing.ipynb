```python
import pandas as pd
import string
from nltk.corpus import stopwords
```

```python
df = pd.read_csv('../data/df_unique.csv')
```

```python
df['text'] = df['text'].apply(lambda x: x.lower())
df
```

```python
df.to_csv('../data/df_text_lower.csv', index=False)
```

```python
import nltk
nltk.download('stopwords')
```

```python
from tqdm import tqdm
tqdm.pandas()

stop_words = set(stopwords.words("english"))
def remove_stopwords(text):
    return " ".join([word for word in text.split() if word not in stop_words])

df['text'] = df['text'].apply(remove_stopwords)
df
```

```python
df.to_csv('../data/df_text_without_stopwords.csv', index=False)
```

```python
nltk.download('punkt')
from nltk.stem import PorterStemmer
from tqdm import tqdm
tqdm.pandas()

ps = PorterStemmer()
def stem_text(text):
    return " ".join([ps.stem(word) for word in nltk.word_tokenize(text)])

df['text'] = df['text'].apply(stem_text)
```

```python
df.to_csv('../data/df_stem_text.csv', index=False)
```
